```python
import os
from ast import literal_eval
import pandas as pd
import sqlite3
```

```python
with open('spiders/ItthonNagyvilagCikkekTartalom_0411_0527.json','r', encoding='utf8') as f:
    article_content = f.read()
```

```python
article_content = literal_eval(article_content)
```

```python
article_content[0]
```

```python
article_df = pd.DataFrame(article_content)

article_df['article_content'] = article_df['article_content'].apply(lambda x: ' '.join(x))
article_df['article_content'] = article_df['article_content'].replace('<[^>]*>', '', regex=True)
article_df['article_title'] = article_df['article_title'].apply(lambda x: ''.join(x))
article_df['article_title'] = article_df['article_title'].replace('<[^>]*>', '', regex=True)
article_df['author'] = article_df['author'].apply(lambda x: ', '.join(x))
article_df['date'] = article_df['date'].apply(lambda x: ''.join(x))
article_df['date'] = article_df['date'].replace('<[^>]*>', '', regex=True)
article_df = article_df.drop_duplicates()
```

```python
conn = sqlite3.connect("origo_articles_2014_2018.db")
```

```python
article_df.to_sql('origo_data_0411_0527', conn)
```
